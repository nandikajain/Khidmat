```python
def change_formatting(word):
    dd,mm,yy = word.split('-')
    
#     #if the date format is already correct then don't swap
    if(int(yy) > 999):
        return yy+"-"+mm+"-"+dd
    return dd+"-"+mm+"-"+yy


change_formatting('16-10-1999')
```

```python
def email_formatting(name,phone,word):
    _,site = word.split('@')
    return name+str(phone)+'@'+site

for i in mylist:
    i[10] = email_formatting(i[2],i[1],i[10])
```
